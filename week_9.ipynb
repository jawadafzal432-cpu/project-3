## initialize the Gemini Client

```python
import os
from google import genai
from google.colab import userdata

os.environ["GEMINI_API_KEY"] = userdata.get('GEMINI_API_KEY')

client = genai.Client()
print('Gemini client initialized successfully!')
```

# Part 1 First API Call

```python
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents='Say hello and introduce yourself!',
    config={
        'max_output_tokens': 150,
        'temperature': 0.7
    }
)

print(response.text)
```

# Part 2 the Conversation Loop

```python
# Initialize a multi-turn chat session with a system instruction
chat_session = client.chats.create(
    model="gemini-2.5-flash",
    config={
        "system_instruction": "You are a helpful assistant. Be friendly, concise, and professional. If you don't know something, say so.",
        "temperature": 0.7,
        "max_output_tokens": 500
    }
)

print('Chatbot ready! Type "quit" to exit.\n')

# Clean conversation loop
while True:
    user_input = input('You: ')

    # Check for exit condition cleanly
    if user_input.lower() == 'quit':
        print('Goodbye!')
        break

    # Send message to the active session
    response = chat_session.send_message(user_input)

    print(f'Assistant: {response.text}\n')
```

## Part 3 - Domain-Specific Assistants

```python
def run_gemini_domain_bot(system_instruction, test_questions=None):
    """
    Helper function to launch a custom-scoped Gemini bot loop.
    """
    chat_session = client.chats.create(
        model="gemini-2.5-flash",
        config={
            "system_instruction": system_instruction,
            "temperature": 0.7,
            "max_output_tokens": 500
        }
    )

    print("="*50)
    print("Gemini Domain Bot Initialized! Type 'quit' to exit.")
    print("="*50)

    if test_questions:
        print("\n--- Running Automated Evaluation Suite ---")
        for q in test_questions:
            print(f"Simulating User Question: {q}")
            res = chat_session.send_message(q)
            print(f"Bot Response: {res.text}\n" + "-"*30)

    print("\n--- Swapping to Interactive Live Session ---")
    while True:
        user_input = input('You: ')
        if user_input.lower() == 'quit':
            print('Exiting Bot Session.\n')
            break
        response = chat_session.send_message(user_input)
        print(f'Assistant: {response.text}\n')
```

#HR Assistant Bot

```python
hr_policy = """You are an HR assistant for a technology company.
Company policies:
Vacation: 15 days per year
Sick leave: Unlimited (with manager approval)
Remote work: 3 days per week
Health insurance: Fully covered
401(k) matching: Up to 5%

Your role:
1. Answer employee questions about policies
2. Be friendly and supportive
3. If unsure, suggest contacting HR department
4. Keep responses concise (2-3 sentences)"""

hr_questions = [
    'How many vacation days do I get?',
    'Can I work from home?',
    'What about health insurance?',
    'How does 401(k) matching work?'
]

run_gemini_domain_bot(hr_policy, hr_questions)
```

# Customer Support Bot

```python
support_policy = """You are a customer support agent for TechShop, an electronics retailer.
Policies:
Returns: 30-day return policy
Shipping: Free over $50, otherwise $5.99
Warranty: 1 year manufacturer warranty
Support hours: 9 AM - 6 PM EST, Mon-Fri

Your tone: Empathetic and patient, Apologize when appropriate, Offer to escalate complex issues, Solution-focused."""

support_scenarios = [
    'I want to return a product I bought 2 weeks ago',
    'How much is shipping?',
    'My laptop stopped working after 6 months'
]

run_gemini_domain_bot(support_policy, support_scenarios)
```
